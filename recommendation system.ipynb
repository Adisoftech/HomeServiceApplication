```python
x = data.groupby('User ID').count()['Rating'] 
x[x]
```

```python
pt = data.pivot_table(index = 'Service Type',columns='User ID', values='Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score
```

```python
def recommend(service_name):
    index = np.where(pt.index == service_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key = lambda x:x[1],reverse=True)[1:6]
    suggestion = []
    for i in similar_items:
        suggestion.append(pt.index[i[0]])
    return suggestion
```

```python
recommend('Bathroom cleaning')
```

```python
services_list = set(data['Service Type'])
services_list
```

```python
result = {}
for service in services_list:
    result[service]=recommend(service)
print("{:<20}\t{:<20}\t{:<20}\t{:<20}\t{:<20}\n".format('Service','Recommendation 1','Recommendation 2','Recommendation 3','Recommendation 4','Recommendation 5'))
for key,value in result.items():
    b1,b2,b3,b4,b5=value
    print("{:<20}\t{:<20}\t{:<20}\t{:<20}\t{:<20}\t{:<20}\n".format(key,b1,b2,b3,b4,b5))
```

```python
import json
with open("sample.json", "w") as outfile: 
    json.dump(result, outfile)
```
